# Missionaries and Cannibals using BFS

```python
from collections import deque

m=int(input('Enter the number of missionaries: '))
c=int(input('Enter the number of cannibals: '))
bc=int(input('Enter the boat capacity: '))

parent=dict()
visited=set()
q=deque()

# left Missionaries, left Cannibals, boat side(0:left, 1:right), right Missionaries, right Cannibals
q.append((m,c,0,0,0))
visited.add((m,c,0,0,0))

while len(q)>0:
    leftM,leftC,boatPos,rightM,rightC=q.popleft()
    
    if (leftM,leftC,boatPos,rightM,rightC)==(0,0,1,m,c):
        break
    
    # boatPos: left
    if boatPos==0:
        for num in range(1,bc+1):
            for miss in range(0,num+1):
                if miss>leftM or num-miss>leftC:
                    continue
                
                lm,lc,rm,rc=leftM-miss,leftC-(num-miss),rightM+miss,rightC+(num-miss)
                
                # valid move
                if (lm==0 or lm>=lc) and (rm==0 or rm>=rc):
                    if (lm,lc,1,rm,rc) not in visited:
                        q.append((lm,lc,1,rm,rc))
                        visited.add((lm,lc,1,rm,rc))
                        parent[(lm,lc,1,rm,rc)]=(leftM,leftC,boatPos,rightM,rightC)
                
    elif boatPos==1:# boatPos: right
        for num in range(1,bc+1):
            for miss in range(0,num+1):
                if miss>rightM or num-miss>rightC:
                    continue
                
                lm,lc,rm,rc=leftM+miss,leftC+(num-miss),rightM-miss,rightC-(num-miss)
                
                if (lm==0 or lm>=lc) and (rm==0 or rm>=rc):
                    if (lm,lc,0,rm,rc) not in visited:
                        q.append((lm,lc,0,rm,rc))
                        visited.add((lm,lc,0,rm,rc))
                        parent[(lm,lc,0,rm,rc)]=(leftM,leftC,boatPos,rightM,rightC)
                
if (0,0,1,m,c) not in parent:
    print('No solution!')
else:
    ans=[]
    curr=(0,0,1,m,c)
    
    while curr in parent:
        ans.append(curr)
        curr=parent[curr]
        
    ans.append(curr)
    
    for item in ans[::-1]:
        print(item)
```

# Missionaries and Cannibals using State Space Search

```python
m=int(input('Enter the number of missionaries: '))
c=int(input('Enter the number of cannibals: '))
bc=int(input('Enter the boat Capacity: '))

visited=set()
path=[]

def StateSpaceSearch(currState):
    visited.add(currState)
    path.append(currState)
    
    if currState==(0,0,1,m,c):
        return True
    
    leftM,leftC,boatPos,rightM,rightC=currState
    
    # boatPos: left
    if boatPos==0:
        for num in range(1,bc+1):
            for miss in range(0,num+1):
                if miss>leftM or num-miss>leftC:
                    continue
                
                lm,lc,rm,rc=leftM-miss,leftC-(num-miss),rightM+miss,rightC+(num-miss)
                
                # valid move
                if (lm==0 or lm>=lc) and (rm==0 or rm>=rc):
                    if (lm,lc,1,rm,rc) not in visited:
                        if StateSpaceSearch((lm,lc,1,rm,rc))==True:
                            return True
                
    elif boatPos==1:# boatPos: right
        for num in range(1,bc+1):
            for miss in range(0,num+1):
                if miss>rightM or num-miss>rightC:
                    continue
                
                lm,lc,rm,rc=leftM+miss,leftC+(num-miss),rightM-miss,rightC-(num-miss)
                
                if (lm==0 or lm>=lc) and (rm==0 or rm>=rc):
                    if (lm,lc,0,rm,rc) not in visited:
                        if StateSpaceSearch((lm,lc,0,rm,rc))==True:
                            return True
                        
    path.pop()
    
if StateSpaceSearch((m,c,0,0,0))==False:
    print('No Solution!')
else:
    for item in path:
        print(item)
```
