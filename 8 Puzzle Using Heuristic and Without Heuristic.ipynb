# 8 Puzzle Problem (Branch and Bound) (Heuristic)

```python
# import heapq
# implement priority queue class
# create node class with overloaded __lt__
# now make a funtion to find the solution using priority queue

import heapq

class PriorityQueue:
    def __init__(self):
        self.arr=[]
        
    def push(self,item):
        heapq.heappush(self.arr,item)
        
    def pop(self):
        return heapq.heappop(self.arr)
    
    def empty(self):
        return len(self.arr)==0
    
class Node:
    def __init__(self,matrix,parent,x,y,level,cost):
        self.matrix=matrix
        self.parent=parent
        self.x=x # x pos of empty cell
        self.y=y # y pos of empty cell
        self.level=level
        self.cost=cost
        
    def __lt__(self,other):
        return self.cost+self.level<other.cost+other.level

def calculateCost(currMatrix,newMatrix):
    
    cost=0
    
    for i in range(len(currMatrix)):
        for j in range(len(newMatrix)):
            if currMatrix[i][j]!=newMatrix[i][j]:
                cost+=1
                
    return cost

def newNode(currNode,newZeroPos):
    
    newMat=[]
    for i in range(len(currNode.matrix)):
        row=[]
        for j in range(len(currNode.matrix)):
            row.append(currNode.matrix[i][j])
        newMat.append(row)
        
    # swap operation
    x1=currNode.x
    y1=currNode.y
    x2=newZeroPos[0]
    y2=newZeroPos[1]
    
    newMat[x1][y1],newMat[x2][y2]=newMat[x2][y2],newMat[x1][y1]
    
    cost=calculateCost(newMat,final)
    
    new_node=Node(newMat,currNode,x2,y2,currNode.level+1,cost)
    
    return new_node
    
dx=[-1,1,0,0]
dy=[0,0,-1,1]

def printPath(currNode):
    
    if currNode==None:
        return
    
    printPath(currNode.parent)
    
    for i in range(len(currNode.matrix)):
        for j in range(len(currNode.matrix)):
            print(currNode.matrix[i][j],end=' ')
        print('')
        
    print('\n')

def solve(initial,zeroPos):
    pq=PriorityQueue()
   
    cost=calculateCost(initial,final)
    currNode=Node(initial,None,zeroPos[0],zeroPos[1],0,cost)
    
    pq.push(currNode)
    
    while not pq.empty():
        
        currNode=pq.pop()
        
        if currNode.cost==0:
            printPath(currNode)
            return
        
        for i in range(4):
            
            newX=currNode.x+dx[i]
            newY=currNode.y+dy[i]
            
            if newX>=0 and newX<dim and newY>=0 and newY<dim:
                new_node=newNode(currNode,(newX,newY))
                
                pq.push(new_node)
    
dim=int(input('Enter the dimension (n for n*n): '))

initial=[]
for i in range(dim):
    row=input('Enter the '+str(i+1)+'th row: ').split()
    for j in range(len(row)):
        row[j]=int(row[j])
        
    initial.append(row)
    
final=[]
for i in range(dim):
    row=[]
    for j in range(1,dim+1):
        row.append(dim*i+j)
    final.append(row)
final[dim-1][dim-1]=0

def checkInputValidity(initial):
    st=set()
    posx,posy=-1,-1
    for i in range(dim):
        for j in range(dim):
            st.add(initial[i][j])
            if initial[i][j]==0:
                posx=i
                posy=j
                
    for i in range(dim*dim):
        if i not in st:
            return False
    
    return posx,posy

zeroPos=checkInputValidity(initial)
if zeroPos==False:
    print('Invalid Input!')
else:
    solve(initial,zeroPos)
```

# 8 Puzzle BFS (Without Heuristic)

```python
from collections import deque
import copy

dim=int(input('Enter the number of dimension(n for n*n): '))

print('Enter the initial state row by row: ')
initial=[]
for i in range(dim):
    row=input().split(' ')
    for i in range(dim):
        row[i]=int(row[i])
    initial.append(row)
    
print('Enter the final state row by row: ')
final=[]
for i in range(dim):
    row=input().split(' ')
    for i in range(dim):
        row[i]=int(row[i])
    final.append(row)
    
class Node:
    def __init__(self,matrix,parent,x,y):
        self.matrix=matrix
        self.x=x
        self.y=y
        self.parent=parent
        
def to_tuple(matrix):
    t=[]
    
    for i in range(len(matrix)):
        t.append(tuple(matrix[i]))
        
    return tuple(t)

def findZeroPos(matrix):
    posX,posY=-1,-1
    
    for i in range(len(matrix)):
        for j in range(len(matrix)):
            if matrix[i][j]==0:
                posX=i
                posY=j
                
    return posX,posY

def printPath(currNode):
    
    if currNode==None:
        return
    
    printPath(currNode.parent)
    
    for i in range(len(currNode.matrix)):
        for j in range(len(currNode.matrix)):
            print(currNode.matrix[i][j],end=' ')
        print('')
        
    print('')

dx=[0,0,-1,1]
dy=[-1,1,0,0]
def solve(initial,final):
    
    visited=set()
    q=deque()
    zeroPos=findZeroPos(initial)
    
    visited.add(to_tuple(initial))
    q.append(Node(initial,None,zeroPos[0],zeroPos[1]))
    
    while len(q)>0:
        
        currNode=q.popleft()
        
        if currNode.matrix==final:
            print('Steps: ')
            printPath(currNode)
            return
        
        for i in range(4):
            newX=currNode.x+dx[i]
            newY=currNode.y+dy[i]
            
            # valid move
            if newX>=0 and newX<dim and newY>=0 and newY<dim:
                new_mat=copy.deepcopy(currNode.matrix)
                
                oldX=currNode.x
                oldY=currNode.y
                # swap operation
                new_mat[oldX][oldY],new_mat[newX][newY]=new_mat[newX][newY],new_mat[oldX][oldY]
                
                t=to_tuple(new_mat)
                if t not in visited:
                    visited.add(t)
                    q.append(Node(new_mat,currNode,newX,newY))
                    
    print('No solution!')
    

solve(initial,final)
```

# Water Jug BFS

```python
from collections import deque

jug1=int(input('Enter the capacity of Jug1: '))
jug2=int(input('Enter the capacity of Jug2: '))

goal=int(input('Enter the goal amount: '))

def printPath(parent,curr):
    
    lst=[]
    while curr in parent:
        lst.append(curr)
        curr=parent[curr]
        
    print((0,0))
    for each in lst[::-1]:
        print(each)

def solve(jug1,jug2,goal):
    
    parent={}
    visited=set()
    q=deque()
    q.append((0,0))
    visited.add((0,0))
    
    while len(q)>0:
        
        curr=q.popleft()
        j1,j2=curr
        
        if j1==goal or j2==goal:
            printPath(parent,curr)
            return
            
        # empty 1st jug
        if (0,j2) not in visited:
            q.append((0,j2))
            visited.add((0,j2))
            parent[(0,j2)]=curr
        
        # empty 2nd jug
        if (j1,0) not in visited:
            q.append((j1,0))
            visited.add((j1,0))
            parent[(j1,0)]=curr
        
        # fill 1st jug
        if (jug1,j2) not in visited:
            q.append((jug1,j2))
            visited.add((jug1,j2))
            parent[(jug1,j2)]=curr
        
        # fill 2nd jug
        if (j1,jug2) not in visited:
            q.append((j1,jug2))
            visited.add((j1,jug2))
            parent[(j1,jug2)]=curr
        
        # pour water from jug1 to jug2
        t=min(j1,jug2-j2)
        if (j1-t,j2+t) not in visited:
            q.append((j1-t,j2+t))
            visited.add((j1-t,j2+t))
            parent[(j1-t,j2+t)]=curr
        
        # pour water from jug2 to jug1
        t=min(j2,jug1-j1)
        if (j1+t,j2-t) not in visited:
            q.append((j1+t,j2-t))
            visited.add((j1+t,j2-t))
            parent[(j1+t,j2-t)]=curr
            
    print('No Solution!')
    
    
solve(jug1,jug2,goal)
```
