```python
from collections import deque

jug1=int(input('Enter the capacity of Jug1: '))
jug2=int(input('Enter the capacity of Jug2: '))

goal=int(input('Enter the goal amount: '))

def printPath(parent,curr):
    
    lst=[]
    while curr in parent:
        lst.append(curr)
        curr=parent[curr]
        
    print((0,0))
    for each in lst[::-1]:
        print(each)

def solve(jug1,jug2,goal):
    
    parent={}
    visited=set()
    q=deque()
    q.append((0,0))
    visited.add((0,0))
    
    while len(q)>0:
        
        curr=q.popleft()
        j1,j2=curr
        
        if j1==goal or j2==goal:
            printPath(parent,curr)
            return
            
        # empty 1st jug
        if (0,j2) not in visited:
            q.append((0,j2))
            visited.add((0,j2))
            parent[(0,j2)]=curr
        
        # empty 2nd jug
        if (j1,0) not in visited:
            q.append((j1,0))
            visited.add((j1,0))
            parent[(j1,0)]=curr
        
        # fill 1st jug
        if (jug1,j2) not in visited:
            q.append((jug1,j2))
            visited.add((jug1,j2))
            parent[(jug1,j2)]=curr
        
        # fill 2nd jug
        if (j1,jug2) not in visited:
            q.append((j1,jug2))
            visited.add((j1,jug2))
            parent[(j1,jug2)]=curr
        
        # pour water from jug1 to jug2
        t=min(j1,jug2-j2)
        if (j1-t,j2+t) not in visited:
            q.append((j1-t,j2+t))
            visited.add((j1-t,j2+t))
            parent[(j1-t,j2+t)]=curr
        
        # pour water from jug2 to jug1
        t=min(j2,jug1-j1)
        if (j1+t,j2-t) not in visited:
            q.append((j1+t,j2-t))
            visited.add((j1+t,j2-t))
            parent[(j1+t,j2-t)]=curr
            
    print('No Solution!')
    
    
solve(jug1,jug2,goal)
```
