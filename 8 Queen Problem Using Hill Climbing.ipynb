# 8 Queen Problem using Hill Climbing

```python
# take input N
# make a state array
# call hillClimbing Function passing this state
# inside this function do below steps in a loop:
# if cost of curr state == 0 then print it and break
# else find the best neighbouring state(having smallest cost) using another function
# if neighbouring state and curr state is same, then make a random jump(because of local optima)
# else make the neighbouring state as curr state for next iteration
```

```python
import random
import copy

N=int(input('Enter N: '))

def printState(state):
    print('')
    for i in range(N):
        for j in range(N):
            if state[j]==i:
                print('X',end=' ')
            else:
                print('O',end=' ')
        print('')

def calculateCost(state):
    count=0
    for i in range(N):
        for j in range(N):
            if state[j]!=i:
                continue
            
            # up
            x,y=i-1,j
            while x>=0:
                if state[y]==x:
                    count+=1
                    break
                    
                x-=1

            # down
            x,y=i+1,j
            while x<N:
                if state[y]==x:
                    count+=1
                    break
                    
                x+=1

            # left
            x,y=i,j-1
            while y>=0:
                if state[y]==x:
                    count+=1
                    break
                    
                y-=1

            # right
            x,y=i,j+1
            while y<N:
                if state[y]==x:
                    count+=1
                    break
                    
                y+=1

            # up left
            x,y=i-1,j-1
            while x>=0 and y>=0:
                if state[y]==x:
                    count+=1
                    break
                    
                x-=1
                y-=1

            # up right
            x,y=i-1,j+1
            while x>=0 and y<N:
                if state[y]==x:
                    count+=1
                    break
                    
                x-=1
                y+=1

            # down left
            x,y=i+1,j-1
            while x<N and y>=0:
                if state[y]==x:
                    count+=1
                    break
                    
                x+=1
                y-=1

            # down right
            x,y=i+1,j+1
            while x<N and y<N:
                if state[y]==x:
                    count+=1
                    break
                    
                x+=1
                y+=1
                
    
    return count//2

def getBestNeighbour(state,cost):
    
    nextState=copy.deepcopy(state)
    nextCost=cost
    
    for i in range(N):
        temp=state[i]
        
        for j in range(N):
            if state[i]==j:
                continue;
                
            state[i]=j
            tempCost=calculateCost(state)
            
            if tempCost<=nextCost:
                nextCost=tempCost
                nextState=copy.deepcopy(state)
                
        state[i]=temp
        
    return nextState,nextCost
                

def nQueenHillClimbing():
    
    currState=[random.randint(0,N-1) for i in range(N)]
    currCost=calculateCost(currState)
    
    while True:
        if currCost==0:
            printState(currState)
            break
            
        nextState,nextCost=getBestNeighbour(currState,currCost)    
        
        if currState==nextState:
            # make a random jump
            currState[random.randint(0,N-1)]=random.randint(0,N-1)
            currCost=calculateCost(currState)
        else:
            currState=nextState
            currCost=nextCost
            
if N==2 or N==3 or N<0:
    print('No Solution!')
else:
    nQueenHillClimbing()
```
