# 2 (a) 3 Missionaries 3 Cannibals

```python
leftM=3
leftC=3
boatPos=0
# 0: left
# 1: right

rightM=0
rightC=0

print('Game Start: ')
print('M : 3\t\t\t\tM : 0')
print('C : 3\t\t\t\tC : 0')
print('Boat: Left')
    
while True:
    print('\nEnter for transfer: ')
    m=int(input('missionaries: '))
    c=int(input('cannibals   : '))
    
    if m+c>2 or m+c==0:
        print('Wrong Input')
        break
        
    if boatPos==0:
        leftM-=m
        rightM+=m
        leftC-=c
        rightC+=c
        boatPos=1
    else:
        leftM+=m
        rightM-=m
        leftC+=c
        rightC-=c
        boatPos=0
        
    print('Current State: ')
    print('M : '+str(leftM)+'\t\t\t\tM : '+str(rightM))
    print('C : '+str(leftC)+'\t\t\t\tC : '+str(rightC))
    
    if boatPos==0:
        print('Boat: Left')
    else:
        print('Boat: Right')
        
    if (leftM>0 and leftC>leftM) or (rightM>0 and rightC>rightM):
        print('You failed. Start again')
        break
    elif rightC==3 and rightM==3:
        print('You Won.')
        break
```

# 2 (b) N Missionaries N Cannibals

```python
leftM=int(input('Enter the number of missionaries: '))
leftC=int(input('Enter the number of cannibals: '))
boatCapacity=int(input('Enter the boat capacity: '))

numMissionaries=leftM
numCannibals=leftC

boatPos=0
# 0: left
# 1: right

rightM=0
rightC=0

print('Game Start: ')
print('M : '+str(leftM)+'\t\t\t\tM : '+str(rightM))
print('C : '+str(leftC)+'\t\t\t\tC : '+str(rightC))
print('Boat: Left')
    
while True:
    print('\nEnter for transfer: ')
    m=int(input('missionaries: '))
    c=int(input('cannibals   : '))
    
    if m+c>boatCapacity or m+c==0:
        print('Wrong Input')
        break
        
    if boatPos==0:
        leftM-=m
        rightM+=m
        leftC-=c
        rightC+=c
        boatPos=1
    else:
        leftM+=m
        rightM-=m
        leftC+=c
        rightC-=c
        boatPos=0
        
    print('Current State: ')
    print('M : '+str(leftM)+'\t\t\t\tM : '+str(rightM))
    print('C : '+str(leftC)+'\t\t\t\tC : '+str(rightC))
    
    if boatPos==0:
        print('Boat: Left')
    else:
        print('Boat: Right')
        
    if (leftM>0 and leftC>leftM) or (rightM>0 and rightC>rightM):
        print('You failed. Start again')
        break
    elif rightC==numCannibals and rightM==numMissionaries:
        print('You Won.')
        break
```
